#Ejercicio 1: Suma de matrices

B = [[12, 10, 8], [6, 4, 2], [0, -2, -4]]

Ejercicio 7: Proyección ortogonal

Dado un vector V = [5, -3, 2] y un vector U = [2, 1, 2], encuentra la proyección ortogonal de V sobre U.

```python
#  EJERCICIO 7: Proyección ortogonal
print("\n== EJERCICIO 7: Proyección de V sobre U ")
V7 = np.array([5, -3, 2])
U7 = np.array([2, 1, 2])
print("Vector V:", V7)
print("Vector U:", U7)
print("Operación: Proyección de V sobre U")
proy = (np.dot(V7, U7) / np.dot(U7, U7)) * U7
print("Resultado:", proy)
```

Ejercicio 8: Producto escalar de proyecciones

Supongamos que tienes tres vectores V = [3, -1, 2], U = [2, 2, -1], y W = [1, 4, -2]. Encuentra el producto escalar de la proyección de V sobre U con la proyección de V sobre W.

```python

# EJERCICIO 8: Producto escalar de proyecciones
import numpy as np
V = np.array([3, -1, 2], dtype=float)
U = np.array([2, 2, -1], dtype=float)
W = np.array([1, 4, -2], dtype=float)

def proj(v, u):
    denom = np.dot(u, u)
    if np.isclose(denom, 0):
        raise ValueError("Vector u es cero.")
    return (np.dot(v, u) / denom) * u

proj_V_on_U = proj(V, U)
proj_V_on_W = proj(V, W)

dot_product = np.dot(proj_V_on_U, proj_V_on_W)

print("Proyección de V sobre U:", proj_V_on_U)
print("Proyección de V sobre W:", proj_V_on_W)
print("\nProducto escalar entre ambas proyecciones:", dot_product)
```

Ejercicio 9: Ortogonalización de Gram-Schmidt

Dado un conjunto de vectores linealmente independientes:

v1 = [1, 1, 0]
v2 = [1, 2, 1]
v3 = [2, 1, 3]

Aplica el proceso de ortogonalización de Gram-Schmidt para obtener un conjunto ortogonal equivalente a los vectores originales.

```python
# EJERCICIO 9: Gram-Schmidt
print("\nEJERCICIO 9: Ortogonalización Gram-Schmidt ")
v1 = np.array([1, 1, 0])
v2 = np.array([1, 2, 1])
v3 = np.array([2, 1, 3])
print("Vectores iniciales:")
print("v1 =", v1)
print("v2 =", v2)
print("v3 =", v3)
print("Operación: Gram-Schmidt")

u1 = v1
u2 = v2 - (np.dot(v2, u1)/np.dot(u1, u1))*u1
u3 = v3 - (np.dot(v3, u1)/np.dot(u1, u1))*u1 - (np.dot(v3, u2)/np.dot(u2, u2))*u2

print("Resultado:")
print("u1 =", u1)
print("u2 =", u2)
print("u3 =", u3)
```
